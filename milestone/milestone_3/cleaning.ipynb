```python
# Read the csv file
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go

#input_file_path ="../data/Data_for_Project.csv " 
#input_file_path ="C:\Users\Lenovo\Documents\GitHub\ELO-Data-Analysis-Project-Level-of-Happiness\data\Data_for_Project.csv" 
df= pd.read_csv("../../data/Data_for_Project.csv", delimiter=';')
df.head()
```

## Data Cleaning

```python
print(df.info())

# Drop rows where all elements are missing
df.dropna(how='all', inplace=True)

# Fill missing values using forward fill method
df.fillna(method='ffill', inplace=True)

# Columns to be converted from string to numeric
columns_to_convert = ['Ladder score', 'Logged GDP per capita', 'Social support', 
                      'Healthy life expectancy', 'Freedom Index', 'The World Giving Index', 
                      'Corruption Perceptions Index']

# Replace commas with dots and convert to numeric
df[columns_to_convert] = df[columns_to_convert].replace({',': '.'}, regex=True)
df[columns_to_convert] = df[columns_to_convert].astype(float)

# Display the first few rows and information about the dataset after cleaning
print(df.head())
print(df.info())

# Save the cleaned data to a new CSV file
output_file_path ="../../data/Cleaned_Data_for_Project.csv" 
#output_file_path = "data/Cleaned_Data_for_Project.csv"
df.to_csv(output_file_path, index=False, sep=';')

print(f"Data has been saved to the file {output_file_path}")
print(df.head())
```
